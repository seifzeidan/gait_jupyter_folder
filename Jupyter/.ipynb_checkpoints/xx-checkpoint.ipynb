```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
import seaborn as sns # for plotting.
import os
import math

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame(columns=['name','mean_LA(x)', 'mean_LA(y)', 'mean_LA(z)', 'mean_GY(x)', 'mean_GY(y)', 'mean_GY(z)'])
```

```python
directory = os.path.join(os.getcwd(), 'straightWalk\\training')
x=1
firstTime = True
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        filepath = os.path.join(directory, filename)
        testOriginal = pd.read_csv(filepath)
        test=testOriginal.copy()
        # Create a new row of data
        new_row = [filepath[55:69],test['LA(x)'].mean(),test['LA(y)'].mean(),test['LA(z)'].mean(),test['GY(x)'].mean(),test['GY(y)'].mean(),test['GY(z)'].mean()]

        # Add the new row to the DataFrame
        df.loc[len(df)] = new_row
```

```python
#Create a svm Classifier
model = svm.SVC(C = 2, # reg paramater
                kernel='sigmoid', #kernel{linear, poly, rbf, sigmoid, precomputed}, default=rbf
               ) # Linear Kernel

#Train the model using the training sets
model.fit(X_training_dataset, y_training_labels)

#Predict the response for test dataset
y_pred = model.predict(X_testing_dataset)
```

```python
from sklearn import metrics
y_pred = model.predict(X_training_dataset)
print("Accuracy:",metrics.accuracy_score(y_training_labels, y_pred))
```
