```python
import pandas as pd
import numpy as np
import os
import math
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python

df = pd.DataFrame(columns=['name','time','GY(x)', 'GY(y)', 'GY(z)'])

filepath = os.path.join(os.getcwd(), 'straightWalk\\ghazaly.csv')
test = pd.read_csv(filepath)
name='ghazaly'
df[['time','GY(x)', 'GY(y)', 'GY(z)']] = test[['TimeStamp','GY(x)', 'GY(y)', 'GY(z)']].copy()
df['name'] = df['name'].fillna(name)
df = df.dropna()
df
```

```python
df['time'] = pd.to_datetime(df['time'], unit='ms')
df
```

```python

# Set the timestamp column as the DataFrame index
df_indexed = df.set_index('time')

# Resample the DataFrame to the desired sampling rate (e.g., 1 second)
resampled_df = df_indexed.resample('5ms').mean()  # You can change '1S' to match your desired sampling rate

resampled_df

```
