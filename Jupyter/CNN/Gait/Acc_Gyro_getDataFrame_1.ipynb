```python
import pandas as pd
import numpy as np
import os
import math
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df_gyro = pd.DataFrame(columns=['name','time','GY(x)', 'GY(y)', 'GY(z)'])

filepath = os.path.join(os.getcwd(), 'straightWalk\\ghazaly.csv')
test = pd.read_csv(filepath)
name='ghazaly'
df_gyro[['time','GY(x)', 'GY(y)', 'GY(z)']] = test[['TimeStamp','GY(x)', 'GY(y)', 'GY(z)']].copy()
df_gyro['name'] = df_gyro['name'].fillna(name)
df_gyro = df_gyro.dropna()
df_gyro
```

```python
df_gyro['time'] = pd.to_datetime(df_gyro['time'], unit='ms')
df_gyro
```

```python

# Set the timestamp column as the DataFrame index
df_indexed_gyro = df_gyro.set_index('time')

# Resample the DataFrame to the desired sampling rate (e.g., 1 second)
resampled_df_gyro = df_indexed_gyro.resample('5ms').mean()  # You can change '1S' to match your desired sampling rate

resampled_df_gyro

```

```python
resampled_df_gyro.reset_index(inplace=True)
resampled_df_gyro
```

```python
df_indexed_acc.reset_index(inplace=True)
df_indexed_acc
```
