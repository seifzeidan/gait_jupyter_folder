```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
import seaborn as sns # for plotting.
import os
import math

from sklearn.preprocessing import LabelEncoder
```

```python
df_train = pd.DataFrame(columns=['name','mean_LA(x)', 'mean_LA(y)', 'mean_LA(z)', 'mean_GY(x)', 'mean_GY(y)', 'mean_GY(z)','sd_LA(x)', 'sd_LA(y)', 'sd_LA(z)', 'sd_GY(x)', 'sd_GY(y)', 'sd_GY(z)'])
df_test = pd.DataFrame(columns=['name','mean_LA(x)', 'mean_LA(y)', 'mean_LA(z)', 'mean_GY(x)', 'mean_GY(y)', 'mean_GY(z)','sd_LA(x)', 'sd_LA(y)', 'sd_LA(z)', 'sd_GY(x)', 'sd_GY(y)', 'sd_GY(z)'])
```

```python
directory = os.path.join(os.getcwd(), 'straightWalk')
firstTime = True
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        filepath = os.path.join(directory, filename)
        testOriginal = pd.read_csv(filepath)
        test=testOriginal.copy()
        
        split_index_20 = int(len(test) * 0.3)
        test_before = test[:-split_index_20]
            
#         test_before= test.copy()
        
        split_index_70 = int(len(test_before) * 0.7)
        X_train = test_before[:split_index_70]
        X_test = test_before[split_index_70:]
        
        new_row_train = [filename[:-4], X_train['LA(x)'].mean(), X_train['LA(y)'].mean(), X_train['LA(z)'].mean(), X_train['GY(x)'].mean(), X_train['GY(y)'].mean(), X_train['GY(z)'].mean(), X_train['LA(x)'].std(), X_train['LA(y)'].std(), X_train['LA(z)'].std(), X_train['GY(x)'].std(), X_train['GY(y)'].std(), X_train['GY(z)'].std()]
        new_row_test = [filename[:-4], X_test['LA(x)'].mean(), X_test['LA(y)'].mean(), X_test['LA(z)'].mean(), X_test['GY(x)'].mean(), X_test['GY(y)'].mean(), X_test['GY(z)'].mean(), X_test['LA(x)'].std(), X_test['LA(y)'].std(), X_test['LA(z)'].std(), X_test['GY(x)'].std(), X_test['GY(y)'].std(), X_test['GY(z)'].std()]

        # Add the new row to the DataFrame
        df_train.loc[len(df_train)] = new_row_train
        df_test.loc[len(df_test)] = new_row_test
```

```python
df_test
```

```python
label_column = 'name'
label_encoder = LabelEncoder()
df_train[label_column] = label_encoder.fit_transform(df_train[label_column])
df_test[label_column] = label_encoder.fit_transform(df_test[label_column])
```

```python
df_train
```

```python
df_train_shuffled = df_train.sample(frac=1)  # frac=1 shuffles all rows
df_train_shuffled = df_train_shuffled.reset_index(drop=True)

df_test_shuffled = df_test.sample(frac=1)  # frac=1 shuffles all rows
df_test_shuffled = df_test_shuffled.reset_index(drop=True)
```

```python
df_train_shuffled
```

```python
X_training_dataset = df_train_shuffled.copy()
X_testing_dataset = df_test_shuffled.copy()

y_training_labels = X_training_dataset.pop('name')
y_testing_labels = X_testing_dataset.pop('name')
```

```python
#Create a svm Classifier
model = svm.SVC(C = 2, # reg paramater
                kernel='linear', #kernel{‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’}, default=’rbf’
               ) # Linear Kernel

#Train the model using the training sets
model.fit(X_training_dataset, y_training_labels)

#Predict the response for test dataset
y_pred = model.predict(X_testing_dataset)

from sklearn import metrics
y_pred = model.predict(X_testing_dataset)
print("Accuracy:",metrics.accuracy_score(y_testing_labels, y_pred))
```
