```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
import seaborn as sns # for plotting.
import os
import math
```

```python
filepath = "D:/online dataset csv files preprocessing/acc_u1-50_w001_Not_Resampled.csv"
acc_all = pd.read_csv(filepath)
filepath = "D:/online dataset csv files preprocessing/gyro_u1-50_w001_Not_Resampled.csv"
gyro_all = pd.read_csv(filepath)
```

```python
acc_all
```

```python
acc_all['name'] = acc_all['name'].str.replace('u', '').str.rstrip('_')
acc_all['name'] = pd.to_numeric(acc_all['name'])
gyro_all['name'] = gyro_all['name'].str.replace('u', '').str.rstrip('_')
gyro_all['name'] = pd.to_numeric(gyro_all['name'])
```

```python
acc_all
```

```python
df_train = pd.DataFrame(columns=['name','mean_LA(x)', 'mean_LA(y)', 'mean_LA(z)', 'mean_GY(x)', 'mean_GY(y)', 'mean_GY(z)','sd_LA(x)', 'sd_LA(y)', 'sd_LA(z)', 'sd_GY(x)', 'sd_GY(y)', 'sd_GY(z)'])
df_test = pd.DataFrame(columns=['name','mean_LA(x)', 'mean_LA(y)', 'mean_LA(z)', 'mean_GY(x)', 'mean_GY(y)', 'mean_GY(z)','sd_LA(x)', 'sd_LA(y)', 'sd_LA(z)', 'sd_GY(x)', 'sd_GY(y)', 'sd_GY(z)'])
for i in range(1,51):
    acc = acc_all[acc_all['name'] == i].copy()
    gyro = gyro_all[gyro_all['name'] == i].copy()
    
    split_index_70 = int(len(acc) * 0.7)
    acc_train = acc[:split_index_70]
    acc_test = acc[split_index_70:]
    
    split_index_70 = int(len(gyro) * 0.7)
    gyro_train = gyro[:split_index_70]
    gyro_test = gyro[split_index_70:]
    
    new_row_train = [i
                     ,acc_train['LAx'].mean(),acc_train['LAy'].mean(),acc_train['LAz'].mean()
                     ,gyro_train['GYx'].mean(),gyro_train['GYy'].mean(),gyro_train['GYz'].mean()
                     ,acc_train['LAx'].std(),acc_train['LAy'].std(),acc_train['LAz'].std() 
                     ,gyro_train['GYx'].std(),gyro_train['GYy'].std(),gyro_train['GYz'].std()]
    new_row_test = [i
                     ,acc_test['LAx'].mean(),acc_test['LAy'].mean(),acc_test['LAz'].mean()
                     ,gyro_test['GYx'].mean(),gyro_test['GYy'].mean(),gyro_test['GYz'].mean()
                     ,acc_test['LAx'].std(),acc_test['LAy'].std(),acc_test['LAz'].std() 
                     ,gyro_test['GYx'].std(),gyro_test['GYy'].std(),gyro_test['GYz'].std()]
    
    df_train.loc[len(df_train)] = new_row_train
    df_test.loc[len(df_test)] = new_row_test
    
```

```python
df_train_shuffled = df_train.sample(frac=1)  # frac=1 shuffles all rows
df_train_shuffled = df_train_shuffled.reset_index(drop=True)

df_test_shuffled = df_test.sample(frac=1)  # frac=1 shuffles all rows
df_test_shuffled = df_test_shuffled.reset_index(drop=True)
```

```python
X_training_dataset = df_train_shuffled.copy()
X_testing_dataset = df_test_shuffled.copy()

y_training_labels = X_training_dataset.pop('name')
y_testing_labels = X_testing_dataset.pop('name')
```

```python
#Create a svm Classifier
model = svm.SVC(C = 2, # reg paramater
                kernel='linear', #kernel{â€˜linearâ€™, â€˜polyâ€™, â€˜rbfâ€™, â€˜sigmoidâ€™, â€˜precomputedâ€™}, default=â€™rbfâ€™
               ) # Linear Kernel

#Train the model using the training sets
model.fit(X_training_dataset, y_training_labels)


from sklearn import metrics
y_pred = model.predict(X_testing_dataset)
print("Accuracy:",metrics.accuracy_score(y_testing_labels, y_pred))
```
