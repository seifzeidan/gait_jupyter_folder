```python
directory = os.path.join(os.getcwd(), 'backpack\\training')
x = 1
firstTime = True
for filename in os.listdir(directory):
    if (filename.endswith(".csv")):
        filepath = os.path.join(directory, filename)
        try:
            testOriginal = pd.read_csv(filepath)
            test = testOriginal.copy()
            num_rows_before = test.shape[0]

            fivePercent = num_rows_before * 0.20
            fivePercent = math.ceil(fivePercent)
            
            real_fivePercent = num_rows_before * 0.1
            real_fivePercent = math.ceil(fivePercent)

            #test = test.iloc[real_fivePercent:]
            test = test.iloc[:-fivePercent]

            num_rows = test.shape[0]
            # Create a new row of data
            new_row = [
                filepath[55:69],
                test['LA(x)'].mean(), test['LA(y)'].mean(), test['LA(z)'].mean(),
                test['GY(x)'].mean(), test['GY(y)'].mean(), test['GY(z)'].mean(),
                test['LA(x)'].std(), test['LA(y)'].std(), test['LA(z)'].std(),
                test['GY(x)'].std(), test['GY(y)'].std(), test['GY(z)'].std()
            ]

            # Add the new row to the DataFrame
            df.loc[len(df)] = new_row
        except Exception as e:
            print(f"Error occurred while reading file: {filename}")
            print(f"Error message: {str(e)}")
            continue
```
