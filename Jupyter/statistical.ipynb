```python

mean_df = pd.DataFrame(columns=['mean'])

# Iterate over the rows in the original DataFrame
for index, row in dfDropClassName.iterrows():
    # Calculate the mean value for each row
    row_mean = row.mean()
    
    # Create a new row in the mean DataFrame with the calculated mean value
    mean_df.loc[index, 'mean'] = row_mean

# Print the mean DataFrame
print(mean_df)
```

```python

sd_df = pd.DataFrame(columns=['sd'])

# Iterate over the rows in the original DataFrame
for index, row in dfDropClassName.iterrows():
    # Calculate the mean value for each row
    row_sd = row.std()
    
    # Create a new row in the mean DataFrame with the calculated mean value
    sd_df.loc[index, 'sd'] = row_sd

# Print the mean DataFrame
print(sd_df)
```

```python

median_df = pd.DataFrame(columns=['median'])

# Iterate over the rows in the original DataFrame
for index, row in dfDropClassName.iterrows():
    # Calculate the mean value for each row
    row_median = row.median()
    
    # Create a new row in the mean DataFrame with the calculated mean value
    median_df.loc[index, 'median'] = row_median

# Print the mean DataFrame
print(median_df)
```
