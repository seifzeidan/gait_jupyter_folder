```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
import seaborn as sns # for plotting.
import os
from sklearn.preprocessing import LabelEncoder
```

```python
data = {'0': [0.696969696969]}
df = pd.DataFrame(data)
```

```python
directory = os.path.join(os.getcwd(), 'straightWalk')
x=1
firstTime = True
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        filepath = os.path.join(directory, filename)
        testOriginal = pd.read_csv(filepath)
        test=testOriginal.copy()
        
        test.drop( axis=1, columns=['GY(x)','GY(y)','GY(z)','TimeStamp'], inplace=True)
        test.dropna( axis=0, subset='LA(x)', inplace=True)
        df_merged2 = pd.melt(test, value_vars=test.columns)
        df_merged2.drop( axis=1, columns=['variable'], inplace=True)
        df_merged2_transposed = df_merged2.transpose()
        df_merged2_transposed['ClassName']=filepath[46:60]
        
        df = pd.concat([df, df_merged2_transposed], axis=0)
        df = df.reset_index(drop=True)
      
        print(df)
        
        
df.to_csv('SWLAO1.csv', index=False)
```
